1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.


```python

average_median_income = df['median_income'].mean()           #This code calculates the average median income
print(f"Average Median Income: {average_median_income}")

plt.figure(figsize=(12, 6))                           #This line initializes a new figure for the plot using Matplotlib.
                                                      #figsize specifies the width and height of the figure in inches.
sns.histplot(df['median_income'], bins=30, kde=True)  #This line uses Seaborn to create a histogram plot.kde adds a kernel density estimate for a smoother representation.bins indicates the number of rectangles into which the data range is divided.
plt.title('Distribution of Median Income')            #This line sets the title of the plot to 'Distribution of Median Income'.
plt.xlabel('Median Income')                           #This line labels the x-axis with 'Median Income'.
plt.ylabel('Frequency')                               #This line labels the y-axis with 'Frequency', indicating the count of occurrences in each bin.
plt.show()                                            #This function is used to display the current figure that you have created
```

  Histograms are commonly used for visualizing the distribution of continuous data(like median_income).This histogram is created using the 'median_income' column. The x-axis(median income) is divided into 30 bins (intervals).The y-axis represents the frequency (count) of 'median_income' values within each bin.The kde adds a kernel density estimate for a smoother representation.
    This histogram is right-skewed it indicate that most people have lower incomes, but there are a few individuals with significantly higher incomes.The spread or variability of the data is generally larger on the right side.Peaks(highest bar in the histogram) in the histogram indicate common income ranges.

3. Show with the help of visualization, how median_income and median_house_values are related?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df, alpha=0.7,edgecolor = "red", facecolor = "Navy")
#alpha=0.5 sets the transparency of the points to avoid overplotting.sets the edgecolor of the points to "red" and facecolor as "navy".
plt.title('Relationship between Median Income and Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

 Scatter plots are commonly used for visualizing the relationship between two continuous variables.Each point on this scatter plot represents a combination of median_income and median_house_value.
The spread, direction, and density of points indicate the nature of the relationship between the two variables. In this scatterplot we can see that a positive linear relationship between median income and median house value, it indicates as the median income of an area increases, the median house value tends to increase as well.
    When people in a certain area have higher median incomes, there's a tendency for houses in that area to have higher median values.
    This relationship is evident on average, individual cases may vary, and other factors can influence house values.we can see some ouliters in this plot it suggests that there are individual data points that deviate significantly from the overall pattern.

6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median_for_columns(df):
  # Select only numeric columns
  numeric_df = df.select_dtypes(include='number')

  median_values = numeric_df.median()
  return median_values

median_result = calculate_median_for_columns(df)
print("Median values for each column:")
print(median_result)
```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
df['total_bedroom_size'] = 'Large'                                  #A new column named 'total_bedroom_size' is added to the DataFrame df.
                                                                    #All rows are initially set to have the value 'Large' in the 'total_bedroom_size' column.
df.loc[df['total_bedrooms'] <= 10, 'total_bedroom_size'] = 'Small'  #Using the loc function, it updates the 'total_bedroom_size' to 'Small' or 'median' based on given condition.
df.loc[(df['total_bedrooms'] >= 11) & (df['total_bedrooms'] < 1000), 'total_bedroom_size'] = 'Medium'


print("DataFrame with the new column 'total_bedroom_size':")
print(df)
```
